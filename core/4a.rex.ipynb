```python
text = "Hello, World"
pattern = "World"
match = re.search(pattern, text)
if match:
    print('True')
else:
    print('False')
```

```python
text = "The quick brown fox jumps over the lazy dog." 
pattern = r"\b\w{4}\b"
re.findall(pattern, text)
```

```python
text = "She sells sea shells by the sea shore." 
pattern = r"sea"
matches = re.finditer(pattern, text)
for match in matches:
    print(match.group(0))
```
